```python
ingredients = ['gin', 'vodka', 'rum']

description_list = list()
for i in ingredients:
    r = requests.get(f"https://www.thecocktaildb.com/api/json/v1/1/search.php?i={i}")

    description = r.json()['ingredients'][0]['strDescription']
    description_list.append({'ingredient':i, 'description':description})
```

```python
from bs4 import BeautifulSoup
```

```python
url = 'https://christophm.github.io/interpretable-ml-book/what-is-machine-learning.html'
r = requests.get(url)
soup = BeautifulSoup(r.text)
```

```python
soup
```

```python
soup.find_all('li')
```

```python
for i in soup.find_all('li')[0:3]:
    print(i.text, end='')
```

```python
text = ''
for i in soup.find_all('li'):
    text += i.text.replace('\n', '')+' '

print(text) 
```

```python
text = text.lower()
print(text[0:500])
text = text.encode('ASCII', errors='ignore').decode()
print(text[0:500])
```

```python
from string import punctuation

print('punctuation libary: ',punctuation,'\n')
for i in punctuation:
    text = text.replace(i, '')
print(text)
```

```python
from nltk.stem import SnowballStemmer

snowball = SnowballStemmer('english')
snowball.stem('running')
```

```python
token = list()

for i in text.split(' '):
    token.append(snowball.stem(i))
new_text = ' '.join(token).strip()
print(new_text)
```

```python
words = list()
for i in text.split(' '):
    words.append(i)
print(words[0:100])
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import pandas as pd
cV = CountVectorizer()
Cv = cV.fit_transform([text])

tF = TfidfVectorizer()
Tf = tF.fit_transform([text])
```

```python
data = pd.DataFrame(Cv.todense(), columns=cV.get_feature_names_out())
data.head().T
```

```python
data_tf = pd.DataFrame(Tf.todense(), columns=tF.get_feature_names_out())
data_tf.head().T
```

```python
import matplotlib.pyplot as plt
token_counts = pd.Series(Cv.toarray().ravel(), index=cV.get_feature_names_out())
token = token_counts.sort_values().tail(10).sort_values()

fig, ax = plt.subplots()
ax.barh(token.index, token.values)
ax.set_xlabel("frequency")
ax.set_ylabel("token")
plt.show()
```
