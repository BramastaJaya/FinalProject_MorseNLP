```python
import pandas as pd
from pathlib import Path
```

```python

source_files = sorted(Path('SPECIL_dataset').glob('*.csv'))

dataframes = []
for file in source_files:
    df = pd.read_csv(file, header=None) 
    df['source'] = file.name
    dataframes.append(df)

df_all = pd.concat(dataframes)
display(df_all)
```

```python
data = df_all.copy()
data.drop(columns=['source',2,3], inplace=True)
```

```python
def clean_lower(text):
    text = text.lower()
    return text
data[0] = data[0].apply(clean_lower)
data[1] = data[1].apply(clean_lower)
data
```

```python
def remove_ascii(text):
    text = text.encode('ASCII', errors='ignore').decode()
    return text

data[0] = data[0].apply(remove_ascii)
data[1] = data[1].apply(remove_ascii)

data
```

```python
import re
#Remove Puncutuation
clean_spcl = re.compile('')
clean_symbol = re.compile('[^0-9a-z]')
def clean_punct(text):
    text = clean_spcl.sub('', text)
    text = clean_symbol.sub(' ', text)
    return text
# Buat kolom tambahan untuk data description yang telah diremovepunctuation   
data[0] = df[0].apply(clean_punct)
data[1] = df[1].apply(clean_punct)

data.head()
```
